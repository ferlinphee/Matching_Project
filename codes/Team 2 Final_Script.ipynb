The code below includes **3** separate algorithms used to run our matching algorithm. The functions are the following:

1.   **Proportion Algorithm:** This function uses proportions to match strings. For this method, we iterated over rows of the left dataset against the flitered right dataset to match values based on our matching function, named calculate_word_match(). This function finds the number of matching words between the left and right combined names and addresses, and then calculates the proportion of the words that match for each row.

2.   **Dynamic Algorithm:** This method incorporates dynamic programming into the calculate_word_match() function. For this method, we create a matrix to compare the words in the left and right strings. We compare each word from the left with each word from the right to fill in this matrix. The output of this function is the proportion of successful matches in the matrix.

3.   **Fuzzywuzzy Algorithm:** This method uses the fuzzywuzzy library to calculate the similarity between two strings based on the Levenshtein distance. This function returns a similarity score between 0 and 100 for each match. It is divided by 100 to get a confidence score of 1.

We also have an additional function called cleaning_func(), which cleans our left and right datasets and returns a tuple of (left,right) cleaned data. We have also included a function that prints our data visualizations, called create_viz().
